```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from scipy.ndimage import gaussian_filter
from PIL import Image
```

```python
import matplotlib.pyplot as plt
import os

row = 1
col = 5  # Assuming you want 5 images per row
plt.figure(figsize=(20, 6))

for folder in train_folders:
    print(f"Checking the folder {folder}")
    folder_path = os.path.join(train_path, folder)
    images = os.listdir(folder_path)[:5]

    # Remove existing axes on the current row (if any)
    for j in range(col):
        plt.gca().remove()  # Assuming you want all previous axes gone

    for i in range(min(len(images), col)):
        image_path = os.path.join(folder_path, images[i])
        image = plt.imread(image_path)
        original_array = np.array(image)
        blurred_array = gaussian_filter(original_array, sigma=1)
        high_pass_filter = original_array - blurred_array
        plt.subplot(row, col, i + 1)
        plt.imshow(high_pass_filter, cmap='gray',)
        plt.title(folder)
```

```python
import matplotlib.pyplot as plt
import os

row = 1
col = 5  # Assuming you want 5 images per row
plt.figure(figsize=(20, 6))

for folder in train_folders:
    print(f"Checking the folder {folder}")
    folder_path = os.path.join(train_path, folder)
    images = os.listdir(folder_path)[:5]

    # Remove existing axes on the current row (if any)
    for j in range(col):
        plt.gca().remove()  # Assuming you want all previous axes gone

    for i in range(min(len(images), col)):
        image_path = os.path.join(folder_path, images[i])
        image = plt.imread(image_path)
        original_array = np.array(image)
        blurred_array = gaussian_filter(original_array, sigma=1)
        high_pass_filter = original_array - blurred_array


        boost_factor = 2
        high_boost_filter = original_array + boost_factor * high_pass_filter

        plt.subplot(row, col, i + 1)
        plt.imshow(high_boost_filter, cmap='gray',)
        plt.title(folder)
```
