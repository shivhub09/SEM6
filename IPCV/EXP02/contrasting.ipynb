```python
import cv2
import numpy as np

# Load the image
img = cv2.imread('ladki.ppm')

# Convert to grayscale (if needed)
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Find minimum and maximum pixel values
min_val, max_val = np.min(gray), np.max(gray)

# Apply contrast stretching
contrast_stretched = np.uint8(255 * ((gray - min_val) / (max_val - min_val)))

# Display the original and contrast-stretched images
cv2.imshow('Original Image', gray)
cv2.imshow('Contrast Stretched Image', contrast_stretched)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
