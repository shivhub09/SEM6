```python
import cv2
import numpy as np
```

```python
def contrast_stretching(image, max , min):
    stretched = 255 * ((image - max) / (max -  min))
    return stretched.astype(np.uint8)
```

```python
def log_transformation(image):
    c = 255 / (np.log(1 + np.max(image)))
    log_transformed = c * np.log(1 + image)
    return log_transformed.astype(np.uint8)
```

```python
def power_law_transformation(image, gamma):
    power_law = np.power(image / 255, gamma) * 255
    return power_law.astype(np.uint8)
```

```python
def enhance_image(image):
    contrast_stretched = contrast_stretching(image, 45, 21)
    log_transformed = log_transformation(image)
    enhanced_image = power_law_transformation(image, 0.4)  # You can adjust the gamma value as needed
    return contrast_stretched, log_transformed, enhanced_image
```

```python
from matplotlib import pyplot as plt
```

```python
def main():
    image = cv2.imread('low1.jpg')
    if image is None:
        print("Error: Unable to read grayscale image.")
        return

    # Read the color image
    color_image = cv2.imread('low1.jpg')
    if color_image is None:
        print("Error: Unable to read color image.")
        return

    plt.imshow(image)

if __name__ == "__main__":
    main()
```

```python
image = cv2.imread('low1.jpg', cv2.IMREAD_GRAYSCALE)

contrast_stretched, log_transformed, enhanced_image = enhance_image(cv2.cvtColor(image, ))
plt.imshow(log_transformed)
plt.imshow(enhanced_image)
```
