```python
import cv2

# Load the image
# Steps to read and show an image :

# store the image path in a variable

# save the image using cv2.imread('path)

# then show the image using cv2.imshow("Image",image)


image_path1 = "image.jpg"  # Replace with the path to your image file
image1 = cv2.imread(image_path1)

# Check if the image is loaded successfully
if image1 is None:
    print(f"Error: Unable to load the image at {image_path1}")
else:
    # Display the image
    cv2.imshow("Image", image1)

    # Wait for a key event and close the window
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt

# Check if the images are loaded successfully
if image1 is None or image2 is None:
    print("Error: Unable to load one or both images")
else:
    # resizing the image
    new_width = 300
    new_height = 300

    resized_image1 = cv2.resize(image1 , (new_width ,new_height))
    resized_image2 = cv2.resize(image2 , (new_width ,new_height))

    # bitwise operation on the image
    bitwise_and = cv2.bitwise_and(resized_image1, resized_image2)
    bitwise_or = cv2.bitwise_or(resized_image1, resized_image2)
    bitwise_not = cv2.bitwise_not(resized_image1, resized_image2)
    bitwise_xor = cv2.bitwise_xor(resized_image1, resized_image2)
    
    # Display the results using Matplotlib
    fig = plt.figure(figsize=(10, 7))

    # Bitwise AND result
    fig.add_subplot(2, 2, 1)
    plt.imshow(cv2.cvtColor(bitwise_and, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title("Bitwise AND")

    fig.add_subplot(2, 2, 2)
    plt.imshow(cv2.cvtColor(bitwise_or, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title("Bitwise OR")


    fig.add_subplot(2, 2, 3)
    plt.imshow(cv2.cvtColor(bitwise_not, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title("Bitwise NOT")
    
    fig.add_subplot(2, 2, 4)
    plt.imshow(cv2.cvtColor(bitwise_xor, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.title("Bitwise XOR")

    plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Assuming image1 and image2 are loaded properly before this code snippet

if image1 is None or image2 is None:
    print("Error: Unable to load one or both images")
else:
    # resizing the images to the same dimensions
    new_width = 300
    new_height = 300

    resized_image1 = cv2.resize(image1, (new_width, new_height))
    resized_image2 = cv2.resize(image2, (new_width, new_height))

    # Check if the images have the same dimensions
    if resized_image1.shape == resized_image2.shape:
        added_image = cv2.add(cv2.cvtColor(resized_image1 , cv2.COLOR_BGR2RGB), cv2.cvtColor(resized_image2 , cv2.COLOR_BGR2RGB))
        
        plt.imshow(added_image)
        plt.axis('off')
        plt.show()
    else:
        print("Error: The dimensions of the images do not match")
```
